```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = '/mnt/data/acc_16.csv'
df = pd.read_csv('/content/drive/MyDrive/Untitled folder/acc_16.csv')

# Display the first few rows of the dataframe
df.head()

```

```python
# Display the column names
df.columns
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

# Handling missing values: Dropping rows with missing values for simplicity
df.dropna(inplace=True)

# Convert data types if needed
# Example: If you have a date column, convert it to datetime
# df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Use 'REL_ROAD' for road conditions
road_conditions_col = 'REL_ROAD'

# Group by road conditions and count the number of accidents
road_conditions = df[road_conditions_col].value_counts()
print(road_conditions)

# Plot road conditions vs number of accidents
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x=road_conditions_col, order=road_conditions.index)
plt.title('Number of Accidents by Road Conditions')
plt.xlabel('Road Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
# Use 'WEATHER' for weather conditions
weather_conditions_col = 'WEATHER'

# Group by weather conditions and count the number of accidents
weather_conditions = df[weather_conditions_col].value_counts()
print(weather_conditions)

# Plot weather conditions vs number of accidents
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x=weather_conditions_col, order=weather_conditions.index)
plt.title('Number of Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
# Use 'HOUR' for the time of day analysis
time_col = 'HOUR'

# Group by hour and count the number of accidents
hourly_accidents = df[time_col].value_counts().sort_index()
print(hourly_accidents)

# Plot time of day vs number of accidents
plt.figure(figsize=(10, 6))
sns.lineplot(data=hourly_accidents)
plt.title('Number of Accidents by Time of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.xticks(range(0, 24))
plt.show()
```

```python
# Assuming 'CF1', 'CF2', 'CF3' are contributing factors
contributing_factors_cols = ['CF1', 'CF2', 'CF3']

for col in contributing_factors_cols:
    if col in df.columns:
        factors = df[col].value_counts()
        print(factors)

        plt.figure(figsize=(10, 6))
        sns.countplot(data=df, x=col, order=factors.index)
        plt.title(f'Number of Accidents by {col}')
        plt.xlabel(col)
        plt.ylabel('Number of Accidents')
        plt.xticks(rotation=45)
        plt.show()
    else:
        print(f"No data available for {col}.")
```
